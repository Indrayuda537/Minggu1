```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ==============================
# 1. BACA CITRA
# ==============================
img = cv2.imread("aglaonema.jpeg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

print("Ukuran citra (HxWxC):", img.shape)

plt.imshow(img_rgb)
plt.title("Citra Asli")
plt.axis("off")
plt.show()

# ==============================
# 2. REPRESENTASI MATRIKS & VEKTOR
# ==============================
print("\nMatriks 5x5 piksel pertama:")
print(img_rgb[:5, :5])

vector = img_rgb.flatten()
print("\nVektor (10 elemen pertama):")
print(vector[:10])

# ==============================
# 3. ANALISIS PARAMETER CITRA
# ==============================
height, width, channel = img.shape
bit_depth = img.itemsize * 8
aspect_ratio = width / height
memory_size_kb = img.nbytes / 1024

print("\n--- Analisis Parameter ---")
print("Resolusi:", width, "x", height)
print("Bit depth:", bit_depth, "bit")
print("Jumlah channel:", channel)
print("Aspect ratio:", round(aspect_ratio,2))
print("Ukuran memori (KB):", round(memory_size_kb,2))

# Jika resolusi 2x dan bit depth setengah
new_width = width * 2
new_height = height * 2
new_bit_depth = bit_depth / 2

new_memory = (new_width * new_height * channel * new_bit_depth) / 8 / 1024
print("Perkiraan memori baru (KB):", round(new_memory,2))

# ==============================
# 4. MANIPULASI CITRA
# ==============================

# Cropping
crop = img_rgb[50:200, 50:200]

# Resizing
resize = cv2.resize(img_rgb, (200, 200))

# Rotasi
rotate = cv2.rotate(img_rgb, cv2.ROTATE_90_CLOCKWISE)

# ==============================
# 5. TAMPILKAN HASIL
# ==============================

plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
plt.title("Asli")
plt.imshow(img_rgb)
plt.axis("off")

plt.subplot(2,2,2)
plt.title("Crop")
plt.imshow(crop)
plt.axis("off")

plt.subplot(2,2,3)
plt.title("Resize")
plt.imshow(resize)
plt.axis("off")

plt.subplot(2,2,4)
plt.title("Rotate")
plt.imshow(rotate)
plt.axis("off")

plt.tight_layout()
plt.show()
```
